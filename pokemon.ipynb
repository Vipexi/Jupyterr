```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
df=pd.read_csv('Pokemon.csv')
```

```python
#df=poke.head(166)
df = df[df['Generation'] < 2] #Taking only the original generation of pokemons
```

```python
df
```

```python
df=df.drop(["Generation"], axis=1) #Dropping the column which I don't find useful
```

```python
df
```

```python
df=df.dropna() #dropping the pokemons with no Type2/NaN
```

```python
df
```

```python
df=df.loc[df['Type 1'] == "Water"] #Interested only in watertypes
```

```python
df=df.sort_values(by="Total",ascending=False) #Finding out which pokemon has the strongest Total
```

```python
df
```

```python
df1=df #Making a new variable to be used later on
```

```python
df.plot(x="Name", y=["Attack", "Speed", "Defense","HP"],kind="bar") #Visualisation parameter is used as "strings" after the kind envelope
plt.show()  #example  figsize=(20,10)
#.area() .bar() .barh().box().hexbin().hist().kde().density().line().pie().scatter() Visualisation tools
```

```python
sum_of_all_values=df[:].sum() #Counting the total of all values
```
