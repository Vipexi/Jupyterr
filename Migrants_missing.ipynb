```python
divided_df=df["Total Dead and Missing"]/3 #Dividing by three because three categories Females,Males and Children
```

```python
print(divided_df)
```

```python
df["Number of Females"]=df["Number of Females"].fillna(divided_df)
```

```python
msno.bar(df)
```

```python
df["Number of Children"]=df["Number of Children"].fillna(divided_df)
```

```python
msno.bar(df)
```
