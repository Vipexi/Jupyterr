```python
import pandas as pd
import numpy as np

array1=np.array([range(0,10),range(10,20),range(20,30)]).T #Using range to get three separate columns 
dataframe=pd.DataFrame(array1, columns=["first_column","second_column","third_column"]) #Making a three columned pandas dataframe.
dataframe 
```

```python
dataframe.iloc[0,0]=997 #row number zero and  first column
dataframe
```

```python
dataframe.iloc[4,0]=1000 #fourth row, first column
dataframe
```

```python
dataframe.iloc[9,1]=998 #ninth row, second columns
dataframe
```

```python
dataframe.iloc[1,2]=999 #first row, third column
dataframe
```

```python
dataframe.iloc[3,0]=-997 #third row, first column
dataframe
```

```python
dataframe.iloc[2,1]=-999 #second row, second column
dataframe
```

```python
dataframe.iloc[6,2]=-998 #sixth row, third column
dataframe
```

```python
dataframe.iloc[8,2]=-1000 #eight row, third column
dataframe
```

```python
dataframe.iloc[3:2,6:2]
dataframe
```
