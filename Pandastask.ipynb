```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)

data = pd.read_csv('titanic_extended.csv')  
data.head(3)
```

```python
data=data.drop(["Name", "Ticket", "Name_wiki", "Body"], axis=1) #Dropping columns of the list
data.head(3) 
```

```python
data=data.rename(columns={"PassengerId" : "ID"}) #Renaming columns
data.head(3)
```

```python
data=data[['ID','Pclass','Sex','Age','Age_wiki','Class','SibSp','Parch','Fare','Cabin','Embarked','Hometown','Boarded','Destination','Lifeboat','WikiId','Survived']]
data.head(3) #Rearranging the columns
```

```python
data.to_csv('Titanic_extended_2.csv',index=False) #Saving the created file
```
