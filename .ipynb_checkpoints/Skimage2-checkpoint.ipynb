```python
import numpy as np
import skimage
from skimage.transform import rescale, resize, downscale_local_mean
from skimage.color import rgb2gray

astronaut = skimage.data.astronaut()
rocket = skimage.data.rocket()

gray=rgb2gray(astronaut)
negative=np.negative(astronaut)
from skimage.filters import threshold_otsu

skimage.io.imshow(astronaut)
skimage.io.show()

gray = rgb2gray(astronaut)

skimage.io.imshow(gray)
skimage.io.show()

thresh = threshold_otsu(gray)
thresh_binary = gray > thresh

skimage.io.imshow(thresh_binary)
skimage.io.show()

astronaut = astronaut[0:427]
rocket = rocket[:,:512]

print(astronaut.shape)
print(rocket.shape)

fusion = np.concatenate((astronaut, rocket), axis=0)
fusion2 = np.concatenate((rocket, astronaut), axis=0)
fusion3 = np.concatenate((fusion, fusion2), axis=1)
skimage.io.imshow(fusion3)
skimage.io.show()
flipped = np.flip(fusion)
skimage.io.imshow(flipped)
skimage.io.show()
skimage.io.imshow(gray)
skimage.io.show()
skimage.io.imshow(rocket)
skimage.io.show()

fusion2 = np.concatenate((rocket, astronaut), axis=1)
skimage.io.imshow(fusion2)
skimage.io.show()

fusion3 = np.concatenate((fusion2), axis=1)
skimage.io.imshow(fusion3)
skimage.io.show()

flippes = np.flip(fusion2)
skimage.io.imshow(flippes)
skimage.io.show
```
