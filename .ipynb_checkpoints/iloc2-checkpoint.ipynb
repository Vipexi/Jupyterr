```python
import pandas as pd
import numpy as np

my_array=np.array([np.arange(0,10),np.arange(10,20),np.arange(20,30)]).T
my_dataframe=pd.DataFrame(my_array, columns=["first_column","second_column","third_column"])
my_dataframe.head(10)
```

```python
my_dataframe.loc[0,"first_column"]=997
my_dataframe.loc[1,"third_column"]=999
my_dataframe.loc[4,"first_column"]=1000
my_dataframe.iloc[3,0]=-997
my_dataframe.iloc[2,1]=-999
my_dataframe.iloc[6,2]=-998
my_dataframe.iloc[8,2]=-1000
my_dataframe.iloc[9,1]=998
my_dataframe.head(10)
```
