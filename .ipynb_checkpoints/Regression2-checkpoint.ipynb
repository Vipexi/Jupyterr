```python
data = {'Year of Birth': [1930,1940,1950,1965,1973,1982,1987,1992,2010], 'Life Expectancy': [59.7,62.9,70.2,69.7,71.4,74.5,75,75.7,78.7]}
```

```python
df=DataFrame(data)
```

```python
df.head(10)
```

```python
polynomiaal_features = PolynomialFeatures(degree=2)
input_dimension_polynomic = polynomiaal_features.fit_transform(df[['Year of Birth']])
regression_tool = linear_model.LinearRegression()
regression_tool.fit(input_dimension_polynomic, df['Life Expectancy'])

plt.title('Year of Birth Vs Life Expectancy', fontsize=14) #import matplotlib.pyplot as visualisation title is the text on the map.
plt.xlabel('Year of Birth', fontsize=14)
plt.ylabel('Life Expectancy', fontsize=14)
plt.grid(True) #makes the squares appear on the map
plt.scatter(df['Year of Birth'], df['Life Expectancy']) # scatters all the data from the two columns along the maps. scatter() and which columns to scatter ([])

regression_curve = regression_tool.predict(input_dimension_polynomic)
#regression_tool.fit(df[['Year of Birth']], df['Life Expectancy'])

#regression_curve = regression_tool.predict(df[['Year of Birth']])
plt.plot(df['Year of Birth'], regression_curve, color='r')

plt.show()
```

```python
regression_tool.predict(polynomiaal_features.fit_transform(np.array([2030]).reshape(-1, 1)))
```
